## 1. Introduction

Worldwide, breast cancer is the most common type of cancer in women and the second most common in terms of mortality. A diagnosis of breast cancer is made when a lump is found (from a self-examination or X-ray) or when a spot is visible on an X-ray. Once a suspicious lump is found, the doctor performs a diagnostic to determine if it is cancerous and, if so, whether it has spread to other parts of the body.

The breast cancer dataset comes from the University of Wisconsin Hospitals in Madison from Dr. William H. Wolberg. 
https://www.kaggle.com/merishnasuwal/breast-cancer-prediction-dataset

## 2. Objectives

In this study, an attempt was made to predict breast cancer using 6 different algorithms:

- Neural Network
- Logistic Regression
- Support Vector Machine (SVM)
- Decision Tree
- Random Forest
- K-Nearest Neighbour

The predictor variable diagnosis - dependent variable (1 = malignant, 0 = benign) used in breast cancer classification has the following characteristics:

- mean_radius (mean distance from center to points on perimeter)
- mean_texture (standard deviation of gray scale values)
- mean_perimeter (mean core tumor size)
- mean_area (mean area of the medullary tumor)
- mean_smoothness (mean local variation in radius length)

A diagnosis of malignant means that the disease is harmful.
The attributes are determined for each cell nucleus.

Source attribute information comes from: 
https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)

Accuracy testing of the 6 different algorithms was measured using misclassifications in the confusion matrix and the area under the ROC curve on the test set.  

The dataset consists of 569 observations and 6 columns. All variables are numerical, there is no missing data issue.

## 4. Models

### 4.0. Splitting the data into a training set and a testing set

To ensure the generalisability of the findings and to avoid overfitting the algorithms, the original dataset was split into a training set and a test set.
- The training set was used to teach the model (80% of observations)
- The testing set was used to test the model (20% of observations)

```python
from sklearn.model_selection import train_test_split
Y = df1['diagnosis']
X = df1.drop(columns=['diagnosis'])
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=9)
```

```python
print('X train shape: ', X_train.shape)
print('Y train shape: ', Y_train.shape)
print('X test shape: ', X_test.shape)
print('Y test shape: ', Y_test.shape)
```

### 4.1. Neural Network

Results of the confusion matrix built on the testing set for the neural network method:

- 92.9% of observations were classified correctly (accuracy)
- 7 zeros were misclassified as ones
- 1 one was incorrectly classified as zero


- 90.1% was the mean of the precision-recall measure, where:
    - precision is defined as the number of true positives relative to the sum of the number of true positives and the number of false positives
    - recall is defined as the number of true positives compared to the sum of the number of true positives and the number of false negatives

Assumed parameters: 
- 3/1 - number of the neurons on a given layer
- activation - relu - corrected linear unit activation function
- activation - sigmoid - sigmoidal function
- optimizer - Adam - adaptive moment estimation algorithm

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train_net = sc.fit_transform(X_train)
X_test_net = sc.fit_transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

#add first input and hidden layer
classifier = Sequential()
classifier.add(Dense(3, activation='relu'))
                                  
#add second hidden layer
classifier.add(Dense(3, activation='relu'))

#adding output layer
classifier.add(Dense(1, activation='sigmoid'))
```

```python
classifier.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
classifier.fit(X_train_net, Y_train, batch_size=100, epochs=100)
```

```python
#Predict test result
Y_predict1 = classifier.predict(X_test_net)
Y_predict1 = (Y_predict1 > 0.5)
```

```python
# The confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns

neural_cm = confusion_matrix(Y_test, Y_predict1)
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(neural_cm, annot=True, linewidth=0.7, linecolor='gray', fmt='g', ax=ax, cmap="YlGnBu")
plt.title('Plot 38 - Neural Network')
plt.xlabel('Y predict')
plt.ylabel('Y test')
plt.show()
```

```python
#Test score
score_neural = (neural_cm[0][0] + neural_cm[1][1])/(neural_cm[0][0] + neural_cm[1][1] + neural_cm[1][0] + neural_cm[0][1])
print(score_neural)
```

```python
#precision and recall
from sklearn.metrics import average_precision_score
average_precision_neural = average_precision_score(Y_test, Y_predict1)

print('Average precision-recall score: {0:0.3f}'.format(
      average_precision_neural))
```
